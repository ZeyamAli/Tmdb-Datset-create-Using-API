#Remove HTML tags

```python
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r' ', text)
```

```python
df_overview = df_overview.apply(remove_html_tags)
```

#Remove URL's

```python
def remove_urls(text):
  pattern = re.compile(r'https?:\\/s+|www\.\s+')
  return pattern.sub(r' ', text)
```

```python
df_overview = df_overview.apply(remove_urls)
```

```python
df_overview[220]
```

#Remove Punctuation

```python

exclude = string.punctuation
```

```python
exclude
```

```python
def remove_punctuation(text):
  for char in exclude:
    text = text.replace(char, ' ')

  return text
```

```python
df_overview = df_overview.apply(remove_punctuation)
```

```python
df_overview[220]
```

#Chat Word Treatement

```python
#Need Dictionary for short wordsor abbreviations for original words
new_text = []
def chat_words(text):
  for w in text.split():
    if w.upper() in chat_words:
      new_text.append((chat_words[w.upper()]))
    else:
      new_text.append(w)

  return ' '.join(new_text)
```

#Removing StopWords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
new_list = []
def remove_stopwords(text):
  for word in text.split():
    if word in stopwords.words('english'):
      new_list.append(" ")
    else:
      new_list.append(word)

  return ' '.join(new_list)
```

```python
df_overview.apply(remove_stopwords)
```
